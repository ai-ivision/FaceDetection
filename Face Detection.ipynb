## Domain: Face Recognition

### Problem Statement:
Nowadays we install web cameras outside & inside our premises to monitor any unusual activities. The system is programmed to correctly detect faces. We have a dataset of face photographs. It contains more than 13,000 images of faces. Each face has been labeled with the name of the person whose image is there. 1680 of the images are of people having two or more distinct photos in the dataset. Every image is centered on a single face & the pixel of each channel (color in RGB) is encoded by a float in the range 0.0 - 1.0. Your work is to study the data and create a face detection model that can detect the face of the person in the image with a bounding box. The analysis includes detecting faces & drawing bounding boxes for detecting the face of the person in the image.

### Input Dataset:
The dataset contains 13,000 images of faces of persons.

### Task to perform:
- **Task1: Data load**
  - Load the data.
- **Task2: Method to detect faces**
  - Function for detecting faces.
- **Task3: Method to draw bounding boxes**
  - Bounding boxes are drawn around the faces.
- **Task4: Load an image**
  - An image is loaded to detect the face.
- **Task5: Display images**
  - Detect faces, draw bounding boxes around faces & display them.

### Import Libraries

```python
import os
import cv2

import matplotlib.pyplot as plt
```

### Loading Directories and Files

```python
dataset_path = "./Data/"
filter_path = "./Data/haarcascade_frontalface_default.xml"
print("Number of sub folder in images folder =",len(os.listdir(dataset_path)))
```

### Defining Necessary Functions

```python
def detect_faces(image):
    face_cascade = cv2.CascadeClassifier(filter_path)
    if face_cascade.empty():
        raise IOError('Could not load Haar cascade classifier xml file.')
    bboxes = face_cascade.detectMultiScale(image, 1.3, 5)
    return bboxes
```

```python
def draw_bounding_boxes(image, bboxes):

    for box in bboxes:
        print(f"Bounding box co-ordinates = {box}\n")
        x1, y1, w, h = box
        cv2.rectangle(image, (x1, y1), (x1+w, y1+h), (0, 255, 0), 4)
```

### Load image

```python
jk_rowling = "./Data/JK_Rowling/"

print("List of images in JK_Rowling folder  :   \n", os.listdir(jk_rowling))
```

### Plot Image

```python
plt.figure(1, figsize=(20, 5))

for index, image in enumerate(os.listdir(jk_rowling)):
    image_path = jk_rowling + image
    print(f"For {index + 1} Image file path = {image_path}\n")

    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    plt.imshow(image)
    plt.axis("off")
    plt.show()
```

### Draw Boxes and Display Image

```python
plt.figure(1, figsize=(20, 5))

for index, image in enumerate(os.listdir(jk_rowling)):
    image_path = jk_rowling + image
    print(f"For {index + 1} Image file path = {image_path}\n")

    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    bboxes = detect_faces(image)

    draw_bounding_boxes(image, bboxes)

    plt.subplot(1, len(os.listdir(jk_rowling)), index +1)
    plt.subplots_adjust(wspace=0.3)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(image)

plt.show()
```
